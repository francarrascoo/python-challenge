### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 4. Productos más y menos vendidos

# 6. Generando gráficos

```python
#Tienda 1

import matplotlib.pyplot as plt

ventas_por_producto = tienda['Producto'].value_counts().head(10)

plt.figure(figsize=(10,6))
ventas_por_producto.plot(kind='bar', color='skyblue')
plt.title('Top 10 Productos más vendidos')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.tight_layout()
plt.show()
```

```python
#Tienda 1

ventas_por_producto = tienda['Producto'].value_counts()

menos_vendidos = ventas_por_producto.tail(10)

plt.figure(figsize=(10,6))
menos_vendidos.plot(kind='bar', color='salmon')
plt.title('Top 10 Productos menos vendidos')
plt.xlabel('Producto')
plt.ylabel('Cantidad de ventas')
plt.tight_layout()
plt.show()
```

```python
#Tienda 1

categoria_counts = tienda['Categoría del Producto'].value_counts()

plt.figure(figsize=(8,8))
categoria_counts.plot(kind='pie', autopct='%1.1f%%', startangle=140)
plt.title('Distribución de Categorías de Producto')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
#Tienda 1
calif_promedio = tienda.groupby('Lugar de Compra')['Calificación'].mean().sort_values(ascending=False)

# Graficar
plt.figure(figsize=(8,5))
plt.plot(calif_promedio.index, calif_promedio.values, marker='o', linestyle='-', color='royalblue')
plt.title('Calificación promedio por tienda')
plt.ylabel('Calificación promedio')
plt.xlabel('Lugar de compra')
plt.ylim(0, 5)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.show()
```
